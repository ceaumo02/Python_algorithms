## Solicitudes a la Web

```python
import requests

response = requests.get('https://www.platzi.com')
response.encoding = 'uft-8'

print(dir(response))
```

```python
print(response.headers)
```

```python
print(response.text)
```

## Extracción de información html

```python
import bs4

soup = bs4.BeautifulSoup(response.text, 'html.parser')

print(soup.title.text)

print(soup.select('meta[name=description]')[0]['content'])

courses_links = soup.select('.RecentCourses-item')
courses = [course.a['href'] for course in courses_links]

for course in courses:
    print(course)
```

## Primera vista de dataset de Noticias

### El Tiempo - Periodico

```python
import bs4
import requests

url = 'https://www.eltiempo.com'

response = requests.get(url)
response.encoding = 'uft-8'

soup = bs4.BeautifulSoup(response.text, 'html.parser')

noticias = soup.select('.title-container a')

for titulo in noticias:
    print(f"Noticia: {titulo.text} | Link: {url}{titulo['href']}")
```

```python
import bs4
import requests

url = 'https://elpais.com/'

response = requests.get(url)
response.encoding = 'uft-8'

soup = bs4.BeautifulSoup(response.text, 'html.parser')

noticias = soup.select('.headline a')

for titulo in noticias:
    print(f"Noticia: {titulo.text} | Link: {url}{titulo['href']}")
```
